```python
import numpy as np
import pandas as pd
from sklearn import *
import datetime as dt
from IPython.core.display import display
```

```python
train_data = pd.read_csv('train')
test_data = pd.read_csv('test')
wea_data = pd.read_csv('weather_data').rename(columns={'calendar_date':'visit_date'})
```

```python
train_data.head()
```

```python
display(train_data.head())
display(train_data.info())
display(test_data.head())
display(test_data.info())
display(wea_data.head())
display(wea_data.info())
```

```python
train = pd.merge(train_data,wea_data,on ='visit_date' ,how='left' )
test = pd.merge(test_data,wea_data,on = 'visit_date',how='left')
```

```python
train
```

```python
train.to_csv('train',index=False)
test.to_csv('test',index=False)
```

```python
test.info()
```

```python
col = [c for c in train if c not in ['id', 'air_store_id', 'visit_date', 'visitors']]
```

```python
X = train[col]
y = pd.DataFrame()
y['visitors'] = np.log1p(train['visitors'])
```

```python
X.head()
```

```python
X = X.values
y = y.values
```

```python
display(X.shape)
display(y.shape)
```

```python
# yarray(252108, 1)rereshape(252108,)  StratifiedKFold
y = y.reshape(252108,)
```

```python
from sklearn.model_selection import StratifiedKFold

skf = StratifiedKFold(n_splits=3, shuffle=False , random_state=None)
print(skf)

for train_index, test_index in skf.split(X, y):
    X_train , X_test = X[train_index] , X[test_index]
    y_train , y_test = y[train_index] , y[test_index]
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from xgboost import XGBRegressor

xgb = XGBRegressor(max_depth=4,n_estimators=50)
xgb.fit(X_train,y_train)
```

```python
pred = xgb.predict(X_test)
```
